```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Percorso locale al modello
model_path = "./qwen-unrestricted-v1"  # <-- cambia con il path reale

# Carica tokenizer e modello
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_path, torch_dtype=torch.float16, device_map="auto", trust_remote_code=True)
model.eval()

# Messaggi in formato chat
messages = [
    {"role": "user", "content": "Ciao, chi sei?"}
]

# Tokenizzazione con chat template (necessario per i modelli Qwen-style)
input_ids = tokenizer.apply_chat_template(
    conversation=messages,
    tokenize=True,
    add_generation_prompt=True,
    return_tensors='pt'
)

# Sposta su GPU se disponibile
input_ids = input_ids.to(model.device)

# Genera risposta
output_ids = model.generate(
    input_ids,
    max_new_tokens=128,
    do_sample=True,
    temperature=0.7,
    top_p=0.95
)

# Decodifica solo la risposta (saltando il prompt)
response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)
print(response)
```
